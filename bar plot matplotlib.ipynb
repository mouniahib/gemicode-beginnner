```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data (you would replace this with your own dataset)
data = {'Role': ['Developer', 'Manager', 'Designer', 'Developer', 'Designer', 'Manager', 'Developer'],
        'Salary': [60000, 80000, 55000, 70000, 60000, 90000, 75000]}

# Create a DataFrame from the sample data
df = pd.DataFrame(data)

# Use .value_counts() to count the occurrences of each role
role_counts = df['Role'].value_counts()

# Create a bar plot
plt.bar(role_counts.index, role_counts.values, color='skyblue')

# Add labels and a title
plt.xlabel('Roles')
plt.ylabel('Count')
plt.title('Distribution of Different Roles')

# Display the plot
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.show()
```
