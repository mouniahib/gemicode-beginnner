```python
import matplotlib.pyplot as plt

# Sample age data
ages = [25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80, 85]

# Create a box plot
plt.boxplot(ages)

# Add labels and a title
plt.xlabel('Ages')
plt.ylabel('Age Distribution')
plt.title('Box Plot of Ages')

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample age data
ages = [25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80, 85]

# Create a box plot with customized box width
sns.boxplot(data=ages, width=0.4, color='skyblue')

# Add labels and a title
plt.xlabel('Ages')
plt.ylabel('Age Distribution')
plt.title('Box Plot of Ages with Customized Width')

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    'Age': [25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80, 85],
    'Salary': [60000, 80000, 55000, 70000, 60000, 90000, 75000, 85000, 70000, 95000, 100000, 110000, 105000]
}

# Create a DataFrame from the sample data
df = pd.DataFrame(data)

# Create a group of box plots
plt.figure(figsize=(8, 6))  # Set the figure size

sns.boxplot(data=df, orient="v", palette="pastel")

# Add labels and a title
plt.ylabel('Values')
plt.title('Box Plots of Age and Salary')

# Display the plot
plt.xticks([0, 1], ['Age', 'Salary'])  # Adding x-axis labels
plt.show()
```
