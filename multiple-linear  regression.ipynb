```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Create a sample dataset
data =pd.read_csv(kc_house_data)


# Extract x and y from the dataset
x = data[["Age", "Seniority"]]
y = data["Salary"].values

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.35, random_state=40)

# Build a linear regression model
model = LinearRegression()

# Fit the model to the training data
model.fit(x_train, y_train)

# Test the model's performance
predicted = model.predict(x_test)

# Calculate and print the Mean Squared Error (MSE) and R-squared (R^2) score
print("MSE:", mean_squared_error(y_test, predicted))
print("R-squared:", r2_score(y_test, predicted))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Create your feature matrix and target vector
x = [[0.44, 0.68], [0.99, 0.23]]
y = [109.85, 155.72]

# Create a LinearRegression model
lg = LinearRegression()

# Create a PolynomialFeatures transformer with a degree of 2
poly = PolynomialFeatures(degree=2)

# Transform the input data using PolynomialFeatures
x_poly = poly.fit_transform(x)

# Fit the linear regression model to the transformed data
lg.fit(x_poly, y)

# Now, you can use the trained model to make predictions for new data
new_data = [[0.75, 0.65]]  # Example new data point

# Transform the new data point using PolynomialFeatures
new_data_poly = poly.transform(new_data)

# Make predictions for the new data point
predicted_value = lg.predict(new_data_poly)

print("Predicted value for new data:", predicted_value)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Assuming you have a dataset with "Age" and "Salary"
x = data["Age"].values.reshape(-1, 1)
y = data["Salary"].values

# Create a PolynomialFeatures transformer with a degree of 2
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)

# Fit a LinearRegression model to the transformed data
lg = LinearRegression()
lg.fit(x_poly, y)

# Generate a range of values for plotting the polynomial regression curve
x_range = np.arange(min(x), max(x), 0.01).reshape(-1, 1)
x_poly_range = poly.transform(x_range)

# Predict the salaries for the range of ages
y_pred = lg.predict(x_poly_range)

# Create a scatter plot of the data points
plt.scatter(x, y, color="red", label="Data Points")

# Plot the polynomial regression curve
plt.plot(x_range, y_pred, color="black", label="Polynomial Regression Curve")

plt.title("Polynomial Regression")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Generate synthetic data for demonstration
np.random.seed(0)
x = np.random.rand(100, 1) * 5
y = 1.5 * x**2 - 2 * x + 1 + np.random.randn(100, 1)

# Create a PolynomialFeatures transformer with a degree of 2
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)

# Fit a LinearRegression model to the transformed data
lg = LinearRegression()
lg.fit(x_poly, y)

# Generate a range of values for plotting the polynomial regression curve
x_range = np.arange(0, 5, 0.1).reshape(-1, 1)
x_poly_range = poly.transform(x_range)

# Predict the values for the range of x values
y_pred = lg.predict(x_poly_range)

# Create a scatter plot of the original data points
plt.scatter(x, y, color="red", label="Data Points")

# Plot the polynomial regression curve
plt.plot(x_range, y_pred, color="blue", label="Polynomial Regression Curve")

plt.title("Polynomial Regression")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
