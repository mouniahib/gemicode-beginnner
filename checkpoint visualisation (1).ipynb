```python
import pandas as pd
df=pd.read_csv('Video games.csv')

```

```python
df.isnull().sum().sum()
```

```python
import matplotlib.pyplot as plt

# Create an area chart for global sales
plt.fill_between(df['Year'], df['Global_Sales'], color='skyblue')
plt.xlabel('Year')
plt.ylabel('Global Sales')
plt.title('Global Sales Trend Over the Years')
plt.show()
```

```python
#For scatter plots showing relationships between different sales regions:

#North American Sales vs. European Sales:


plt.scatter(df['NA_Sales'], df['EU_Sales'], alpha=0.5, color='green')
plt.xlabel('North American Sales')
plt.ylabel('European Sales')
plt.title('North American Sales vs European Sales')
plt.show()
```

```python
#Japanese Sales vs Other Sales:


plt.scatter(df['JP_Sales'], df['Other_Sales'], alpha=0.5, color='blue')
plt.xlabel('Japanese Sales')
plt.ylabel('Other Sales')
plt.title('Japanese Sales vs Other Sales')
plt.show()
```

```python
#Create a Stacked Area Chart for Sales by Platform Over the Years:

#the groupby and stackplot functions in Matplotlib to create a stacked area chart


# Group the data by 'Year' and 'Platform' and calculate the sum of 'Global_Sales' for each combination
platform_sales = df.groupby(['Year', 'Platform'])['Global_Sales'].sum().unstack().fillna(0)

# Get a list of unique years for the x-axis
years = platform_sales.index

# Get a list of platform names for the legend
platforms = platform_sales.columns

# Create a stacked area chart
plt.stackplot(years, platform_sales.values.T, labels=platforms)

# Customize the plot
plt.xlabel('Year')
plt.ylabel('Global Sales')
plt.title('Sales by Platform Over the Years')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))  # Adjust the legend position
plt.margins(0, 0)  # Remove margins to ensure the plot covers the entire area

# Show the plot
plt.show()
```

```python
#Create a Bar Chart for the Top 10 Publishers by Global Sales:

# Pandas for data aggregation and Matplotlib for visualization.


top_publishers = df.groupby('Publisher')['Global_Sales'].sum().nlargest(10)
top_publishers.plot(kind='bar', color='purple')
plt.xlabel('Publisher')
plt.ylabel('Global Sales')
plt.title('Top 10 Publishers by Global Sales')
plt.show()
```

```python
#Create a Pie Chart for Genre Distribution by Global Sales:

#Use Pandas to group and aggregate data by genre and then create a pie chart.


genre_sales = df.groupby('Genre')['Global_Sales'].sum()
genre_sales.plot(kind='pie', autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightblue', 'lightgreen', 'lightsalmon'])
plt.title('Genre Distribution by Global Sales')
plt.ylabel('')
plt.show()
```
